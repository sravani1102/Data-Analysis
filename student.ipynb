```python
import numpy as np
import pandas as pd
```

```python
cse=pd.read_csv('CSE.csv')
it=pd.read_csv('IT.csv')
st=pd.read_csv('STUDENT.csv')
```

```python
#1.Combine the CSE & IT data and display the data
cse.rename(columns={'Professional Elective':'PE'},inplace=True)
pd.merge(cse,it,how='outer')
```

```python
#2.Display all CSE students’ marks along with personal information.
pd.merge(cse,st,on='Regd.No')
```

```python
#3.Print all students Regd.No, Name and professional elective.
'''df=pd.merge(st,cse)[['Regd.No','Name','PE']]
df1=pd.merge(st,it)[['Regd.No','Name','PE']]
pd.merge(df,df1,how='outer')
'''
df=pd.merge(cse,it,how='outer')
df1=pd.merge(st,df)
df1[['Regd.No','Name','PE']]
```

```python
#4.Identify the students whose DAA marks are >18.
df1[df1['DAA']>18]['Regd.No']
```

```python
#5.Display the names and EAMCET ranks of the students who got minimum 12 marks in all courses.
df1[(df1['CN']>=12) & (df1['DAA']>=12) & (df1['AFL']>=12) & (df1['OE']>=12) &(df1['PE']>=12)][['Name','EAMCET RANK']]
```

```python
#6.Calculate mean value of all the subject’s marks.
df1['mean']=df1[['CN','DAA','AFL','OE','PE']].mean(axis=1)
df1[['Regd.No','mean']]
```

```python
#7.Display the names common in both CSE & IT along with Regd.No.
df=pd.merge(st,cse)
df2=pd.merge(st,it)
pd.merge(df,df2,on='Name')['Name']
```

```python
#8.Fill the missing values of the data with  average marks of the subject of specific group.
mean=cse[['CN','DAA','AFL','OE','PE']].mean()
cse.fillna(mean)
```

```python
#9.Divide the students into 5 groups based on average marks.
df1['group']=pd.qcut(df1['mean'],5,labels=False)
df1[['Regd.No','group']]
```

```python
#10.Create  equal sized groups of students based on EAMCET Rank.
df1['group1']=pd.qcut(df1['EAMCET RANK'],6,labels=False)
df1[['Regd.No','group1']]
```
