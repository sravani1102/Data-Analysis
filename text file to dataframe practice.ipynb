```python
import pandas as pd
```

```python
#read_csv can convert text file into dataframe without and delimter
df1=pd.read_csv('practice1.csv')
df1
```

```python
#read_table does require delimiter to convert text file into dataframe
df=pd.read_table('practice1.csv',sep=',')
df
```

```python
#header = None does not takes the values in dataframe as header(takes default values)
df2=pd.read_csv('practice1.csv',header=None)
df2
```

```python
#header values
df2=pd.read_csv('practice1.csv',names=["S.no","Redg_no","Name"])
df2
```

```python
#columns to index_col
df2=pd.read_csv('practice1.csv',names=["S.no","Redg_no","Name"],index_col='S.no')
df2
```

```python
#open csv text file
list(open('practice1.csv'))
```

```python
df=pd.read_table('practice1.csv',sep='\s+')
df
```

```python
df=pd.read_csv('practice1.csv',skiprows=[1,4,8],na_values=['NULL'])
print(df)
```
